```python
# 로또번호 출력하기
win = Tk()                               # 창 생성

win.geometry("600x100")                  # 창 크기 확인
win.title("현재시각 출력하기")           # 창 이름
win.option_add("*font","궁서 20")        # 폰트 옵션 추가

ent = Entry(win)                         # 입력창 투입
ent.pack()                               # 입력창 배치

def lotto_p():
    import requests                                # 리퀘스트 호출
    from bs4 import BeautifulSoup                 # 뷰티플 수프 호출

    url = 'https://m.dhlottery.co.kr/gameResult.do?method=byWin' # url 지정
    req = requests.get(url)                        # url 내부값 가져오기
    soup = BeautifulSoup(req.text,"html.parser")   # 내부값 정렬하기
    txt = soup.find('div',attrs = {                # 수프로 특정 함수를 추출
        'class','bx_lotto_winnum'}).get_text()     # 로또번호 텍스트 추출
    num_list = txt.split('\n')[1:7]                # 로또번호만 따로 사용
    bonus_list = txt.split('\n')[8]                # 보너스 번호 추출

    print("당첨번호")
    print(num_list)
    print("보너스번호")
    print(bonus_list)

btn = Button(win)                        # 버튼 추가
btn.config(text = "로또 번호")           # 버튼 텍스트
btn.config(width = 30)                   # 버튼 넓이
btn.config(command = lotto_p)            # 버튼 실행
btn.pack()                               # 버튼 배치

win.mainloop()                           # 창 실행
```

```python
# 위젯 배치 방법
# 팩은 순차적으로 배치, 그리드는 격자무늬로 배치, place는 어떤 장소에 놓기
# 서로 같이 못씀

# 팩 : 상하좌우로 설정, 다만 패드를 통해서만 위젯사이 공간을 줄 수 있음
# 그리드 : 빈공간을 줄 수 없음. 엑셀처럼 표 형식임 span으로 합치기
# 플레이스 : 절대위치, 상대위치(rel)를 지정 가능

# 버튼 순간이동
win = Tk()                               # 창 생성

win.geometry("400x200")                 # 창 크기 확인
win.title("temp")                        # 창 이름
win.option_add("*font","맑은고딕 25")    # 폰트 옵션 추가

ent = Entry(win)
ent.pack(side = 'top')

btn = Button(win, text='버튼')           # 버튼 추가

def aa():
    btn.pack(side=ent.get())

btn.config(command = aa)
btn.pack(side = 'bottom')                               # 버튼 배치

win.mainloop()                           # 창 실행
```

```python
from tkinter import *
from datetime import datetime
import random

# exe 파일로 만들기
win = Tk()                               # 창 생성

win.geometry("550x150")                 # 창 크기 확인
win.title("aim_game")                        # 창 이름
win.option_add("*font","맑은고딕 20")    # 폰트 옵션 추가

# 라벨
lab = Label(win)
lab.config(text = '표적개수')
lab.grid(column = 0,row = 0, padx = 20 ,pady = 20)

# 엔트리
ent = Entry(win)
ent.grid(column = 1,row = 0, padx = 20 ,pady = 20)

k = 1
def cc():
    global k
    if k < num_t: 
        k +=1
        btn.destroy()
        ran_btn()
    else:
        fin = datetime.now()
        dif_sec = (fin - start).total_seconds()
        btn.destroy()
        lab = Label(win)
        lab.config(text = 'Clear'+str(dif_sec)+'초')
        lab.pack(pady=230)
        

def ran_btn():
    global btn
    btn = Button(win)
    btn.config(bg = 'red')
    btn.config(command = cc)
    btn.config(text = k)
    btn.place(relx = random.random(),rely = random.random())
    
def btn_f():
    global num_t
    global start
    num_t = int(ent.get())
    for wg in win.grid_slaves(): 
        wg.destroy()
    win.geometry("500x500")
    ran_btn()
    start = datetime.now()
# 버튼
btn = Button(win)
btn.config(text = '시작')
btn.config(command = btn_f)
btn.grid(column = 0, row = 1, columnspan=2)

win.mainloop()                           # 창 실행
```

```python
# 리스트 박스
win = Tk()                               # 창 생성

win.geometry("500x500")                 # 창 크기 확인
win.title("listbox")                        # 창 이름
win.option_add("*font","맑은고딕 25")    # 폰트 옵션 추가

# list box
lb = Listbox(win)
lb.config(selectmode='multiple')
lb.insert(0,'1번')
lb.insert(1,'2번')
lb.insert(2,'3번')
lb.insert(3,'4번')
lb.pack()

# button
btn = Button(win)
btn.config(text = '옵션선택')

def click():
    text = lb.curselection()
    print(text)

btn.config(command = click)
btn.pack()

# label
lab = Label(win)
lab.pack()

win.mainloop()                           # 창 실행
```
