```python
from tkinter import Tk
from datetime import datetime
import pyautogui
import time

global_list = [0]

def update_label():
    now = datetime.now()
    formatted_time = now.strftime("%Y-%m-%d %H:%M:%S")
    lab4.config(text=formatted_time)
    x, y = pyautogui.position()
    lab3.config(text=f"현재좌표 x = {x}, y = {y}")
    lab4.after(100, update_label)


win = Tk()
win.geometry("400x500")
win.title("timetest")
win.option_add("*font", "맑은고딕 15")

# Listbox
lb = Listbox(win)
lb.pack()

# Button
btn1 = Button(win)
btn1.config(text='실행옵션')

def click_activate():
    text = lb.pack()
    print(text)

btn1.config(command=click_activate)
btn1.pack()

# Button2
btn2 = Button(win)
btn2.config(text='작업추가')

def click_add_lb():
    lb.insert(global_list[0], f'{global_list[0]+1}번')
    global_list[0] += 1

btn2.config(command=click_add_lb)
btn2.pack()

# Button3
btn3 = Button(win)
btn3.config(text='작업삭제')
def click_del_lb():
    selected_indices = lb.curselection()  # 선택된 항목의 인덱스 가져오기
    lb.delete(selected_indices)  # 선택된 항목을 삭제
    global_list[0] -= 1
btn3.config(command=click_del_lb)  # 삭제 함수 연결

btn3.pack()

# Label
lab = Label(win)
lab.pack()

lab3 = Label(win)
lab3.pack(side='bottom')

lab4 = Label(win)
lab4.pack(side='bottom')

# 최초 실행
update_label()

win.mainloop()
```

```python
# 콤보박스
from tkinter.ttk import *
win = Tk()                               # 창 생성

win.geometry("500x500")                 # 창 크기 확인
win.title("combobox")                        # 창 이름
win.option_add("*font","맑은고딕 25")    # 폰트 옵션 추가

# combo box
cb_list = ['마우스 좌클릭','마우스 우클릭','마우스 더블클릭', '마우스 드래그', '마우스 드래그 우클릭1', '마우스 드래그 우클릭2']
cb = Combobox(win)
cb.config(values = cb_list)
cb.pack()

# button
btn = Button(win)
btn.config(text = '옵션선택')

def click():
    text = cb.get()
    lab.config(text = text)

btn.config(command = click)
btn.pack()

# label
lab = Label(win)
lab.pack()

win.mainloop()                           # 창 실행
```

```python
import time
import threading
from pynput.mouse import Button,Controller
from pynput.keyboard import Listener,KeyCode
```

```python
delay = 0.1
button = Button.left
start = KeyCode(char='1')
stop = KeyCode(char='2')
exitkey = KeyCode(char='3')
mouse = Controller()

class ClickMouse(threading.Thread):
    def __init__(self,delay,button):
        super(ClickMouse, self).__init__()
        self.delay = delay
        self.button = button
        self.running = False
        self.program_running = True
    
    def start_clicker(self):
        self.running = True
        
    def stop_clicker(self):
        self.running = False
    
    def exit(self): 
        self.stop_clicker()
        self.program_running = False
    
    def run(self):
        while self.program_running:
            while self.running:
                mouse.click(self.button)
                time.sleep(self.delay)
            time.sleep(0.1)

click_thread = ClickMouse(delay,button)
click_thread.start()

def on_press(key):
    if key == start:
        click_thread.start_clicker()
        print('AutoMouse Start!')
    elif key == stop:
        click_thread.stop_clicker()
        print("AutoMouse Stop!")
    elif key == exitkey:
        print("Program Out!")
        click_thread.exit()
        listener.stop()
    
with Listener(on_press=on_press) as listener:
    listener.join()
```
